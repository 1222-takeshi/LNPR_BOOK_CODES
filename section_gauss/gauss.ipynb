```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def gauss_1d(x, mu, sigma):
    coef = 1/math.sqrt(2*math.pi*sigma**2)
    exp = math.exp(-1/(2*sigma**2)*(x - mu)**2)
    return coef*exp
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
xs = np.linspace(-50,50,100)
p = [gauss_1d(x,1,10) for x in xs]
ax.plot(xs,p)
```
