```python
import sys  ###logger（下まで全部）
sys.path.append('../scripts/')
from robot import *
```

```python
class LoggerAgent(Agent):
    def __init__(self, nu, omega, interval_time, init_pose):
        super().__init__(nu, omega)
        self.interval_time = interval_time
        self.pose = init_pose
        self.time = 0.0
        
        self.log = open("log.txt", "w")
        
    def decision(self, observation):
        self.log.write("action {} {} {} {}\n".format(self.time, *self.pose))
        self.log.flush()
        
        for lm in observation:
            self.log.write("obs {} {} {} {}\n".format(self.time, lm[1], *lm[0]))
            self.log.flush()
        
        self.pose = IdealRobot.state_transition(self.nu, self.omega, self.interval_time, self.pose)
        self.time += self.interval_time
        
        return self.nu, self.omega
```

```python
if __name__ == '__main__': 
    time_interval = 1
    world = World(40, time_interval, debug=False) 

    ###真の地図を作成###
    m = Map()
    landmark_positions = [(-4,2), (2,-3), (3,3), (0,4), (1,1), (-3,-1)]
    for p in landmark_positions:
        m.append_landmark(Landmark(*p))
        
    world.append(m) 

    ### ロボットを作る ###
    init_pose = np.array([0,0,0]).T
    a = LoggerAgent(0.2, 10.0/180*math.pi, time_interval, init_pose)
    r = Robot(init_pose, sensor=Camera(m), agent=a, color="red")
    world.append(r)

    world.draw()
```
