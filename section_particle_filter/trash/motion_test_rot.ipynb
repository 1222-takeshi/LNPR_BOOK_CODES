```python
import sys     ###motion_test_header
sys.path.append('../scripts/')
from robot import *
```

```python
import copy    ###motion_test_linear

world = World(40.0, 0.1)  

initial_pose = np.array([0, 0, 0]).T
robots = []
r = Robot(initial_pose, sensor=None, agent=Agent(0.0, 0.1)) 

for i in range(100):
    copy_r = copy.copy(r)
    copy_r.distance_until_noise = copy_r.noise_pdf.rvs()
    world.append(copy_r)
    robots.append(copy_r)

world.draw()
```

```python
import pandas as pd ###motion_test_stats_rot
poses = pd.DataFrame([ [math.sqrt(r.pose[0]**2 + r.pose[1]**2), r.pose[2]] for r in robots], columns=['r', 'theta'])
poses.transpose()
```

```python
print(poses["theta"].var())
print(poses["theta"].mean())
math.sqrt(poses["theta"].var()/poses["theta"].mean())
```
