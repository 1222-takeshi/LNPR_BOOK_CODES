```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gamma, norm
```

```python
def gen_lambda_dist(a, b):
    return gamma(a, scale=b)

def gen_mu_dist(mu_0, beta, lmd):
    return norm(loc=mu_0, scale=np.sqrt(1/lmd))
```

```python
lambda_dist = gen_lambda_dist(1.0, 0.001)

xs = np.arange(0, 0.01, 0.00001)
ys = [lambda_dist.pdf(x) for x in xs]
plt.plot(xs, ys)
plt.show()
```

```python
lmd = lambda_dist.rvs()
mu_dist = gen_mu_dist(200, 1, lmd)
print(np.sqrt(1/lmd))

xs = np.arange(-100, 500, 0.1)
ys = [ mu_dist.pdf(x) for x in xs]
plt.plot(xs, ys)
plt.show()
```
